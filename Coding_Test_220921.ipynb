## 미로의 최단거리 통로(BFS 활용)
- 7*7 격자판 미로를 탈출하는 최단경로의 경로수를 출력하는 프로그램을 작성하세요. 경로수는 출발점에서 도착점까지 가는데 이동한 횟수를 의미한다. 
- 출발점은 격자의 (1, 1) 좌표이고, 탈 출 도착점은 (7, 7)좌표이다. 격자판의 1은 벽이고, 0은 도로이다.
격자판의 움직임은 상하좌우로만 움직인다. 미로가 다음과 같다면
- 위와 같은 경로가 최단 경로이며 경로수는 12이다.

<hr>

- ▣ 입력설명
    - 7*7 격자판의 정보가 주어집니다.
- ▣ 출력설명
    - 첫 번째 줄에 최단으로 움직인 칸의 수를 출력한다. 도착할 수 없으면 -1를 출력한다.

### 참고 풀이

```python
'''
(0,0)이 큐로 들어가서 4방향으로 뻗을 수 있음 
    4방향 중 한번에 갈 수 있는 부분만 이동하면서 +1 
        이동한 부분에서도 4방향으로 뻗어나감 
            위와 같은 방법으로 계속 진행하는데 한번 온 곳은 체크해줘야함(돌아가지 못하게) 
    (6,6) 좌표의 값을 출력해줌 
    BFS로 구한 값이 최단 거리임 
'''
```

```python
from collections import deque
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]
board = [list(map(int, input().split())) for _ in range(7)]
dis = [[0] * 7 for _ in range(7)]

Q = deque()
Q.append((0,0))
board[0][0] = 1 
# 한번 방문했으면 1로 바꿔줌 

while Q: 
    # Q에 뭐가 있으면 반복 진행 
    tmp = Q.popleft()
    
    for i in range(4):
        # 4방향으로 가닥을 뻗기 때문 
        x = tmp[0] + dx[i]
        y = tmp[1] + dy[i]

        if 0 <= x <= 6 and 0<=y<=6 and board[x][y] == 0:
            # 정해진 틀안에 있어야하고 0일 떄만 갈 수 있음 
            board[x][y] = 1 
            # check 

            dis[x][y] = dis[tmp[0]][tmp[1]]+1 
            # 값을 계속 더해주면서 진행하게 되면 6,6일 때 그 값을 출력하면 끝나기 때문에 +1 
            Q.append((x,y))

# 도달하지 못했다면 
if dis[6][6] == 0:
    print(-1)

else:
    print(dis[6][6])


# Q라는 곳에 값이 있어야 반복문이 돌아가는데 그 반복문이 돌아가기 위해서는 특정 조건을 만족해야함 
```

## 미로탐색(DFS)
- 7*7 격자판 미로를 탈출하는 경로의 가지수를 출력하는 프로그램을 작성하세요. 출발점은 격 자의 (1, 1) 좌표이고, 탈출 도착점은 (7, 7)좌표이다.

### 참고 풀이

```python
'''
가지수를 구하는 방법의 수를 구하는 문제이므로 DFS 
DFS는 한 곳으로 쭉 뻗어나가서 도착점이 도착하는 것을 의미한다고 생각해보자 

0일 때 계속 쭉 뻗어나가면서 check 진행  도착점에 도착한다고 count 진행 
back을 진행하는데 check했던 것을 풀어줘야함 <- 이게 나랑 다르네!!! 
'''
```

```python
dx = [-1, 0,1,0]
dy = [0, 1,0,-1]

def DFS(x,y):
    global cnt 

    if x == 6 and  y == 6:
        cnt += 1 

    else:
        for i in range(4):
            # 4방향으로 뻗어나가야 하기 때문에 
            xx = x + dx[i]
            yy = y + dy[i]
            # xx, yy는 앞으로 갈 방향을 뜻함 

            if 0<= xx <= 6 and 0<= yy <= 6 and board[xx][yy] == 0:
                board[xx][yy] = 1 
                DFS(xx, yy)

                # 경우의 수를 구하는 것은 back해서 되돌아가야하기 때문에 check를 풀어줘야함 
                board[xx][yy] = 0 
                # 통로를 만들어줘서 다른 길을 만듬 
                # 난 저걸 안적어서 안된건감 
```

## 등산경로(DFS)
- 등산을 매우 좋아하는 철수는 마을에 있는 뒷산에 등산경로를 만들 계획을 세우고 있습니다. 
- 마을 뒷산의 형태를 나타낸 지도는 N*N 구역으로 나뉘어져 있으며, 각 구역에는 높이가 함께 나타나 있습니다.
N=5이면 아래와 같이 표현됩니다.

- 어떤 구역에서 다른 구역으로 등산을 할 때는 그 구역의 위, 아래, 왼쪽, 오른쪽 중 더 높은 구역으로만 이동할 수 있도록 등산로를 설계하려고 합니다. 
- 등산로의 출발지는 전체 영역에서 가장 낮은 곳이고, 목적지는 가장 높은 곳입니다. 출발지와 목적지는 유일합니다. 지도가주어지면출발지에서도착지로갈수있는 등산경로가몇가지인지구하는프로그 램을 작성하세요.

<hr>

- ▣ 입력설명
    - 첫 번째 줄에 N(5<=N<=13)주어지고, N*N의 지도정보가 N줄에 걸쳐 주어진다.

- ▣ 출력설명
    - 등산경로의 가지수를 출력한다

```python
'''
현재 있는 곳보다 큰 값이 있는 곳으로 가도록 진행 
최종 100을 만나게 된다면 종료 
'''
```

```python
N = int(input())
li = []
for i in range(N):
    a = list(map(int, input().split()))
    li.append(a)
```

```python
for i in li:
    print(i)
```

```python
dx = [-1,0,1,0]
dy = [0,1,0,-1]
cnt = 0 

def DFS(x, y):
    global cnt 
    if li[x][y] == 100:
        cnt += 1 

    else:
        for i in range(4):
            xx = x + dx[i]
            yy = y + dy[i]

            if 0<= xx < N and 0<= yy < N and li[x][y] <= li[xx][yy]:
                # print(xx,yy)
                DFS(xx, yy)
```
